```python
import mne
import pandas as pd
import matplotlib.pyplot as plt

# Path to the .edf file
file_path = r"D:\Inje University Work\Publications Inje Univ\EEG Workload and Not Workload\Epilipsy\aaaaaanr\s001_2003\02_tcp_le\aaaaaanr_s001_t001.edf"

# Load the EDF file
raw = mne.io.read_raw_edf(file_path, preload=True)

# Display a summary of the data
print(raw.info)

# Pick the first 21 channels
raw_21 = raw.copy().pick_channels(raw.ch_names[:21])

# Extract data and convert to pandas DataFrame
data, times = raw_21.get_data(return_times=True)
channel_names = raw_21.ch_names
fs = 256  # Sampling frequency

# Define a 4-second window
window_size = 400 * fs
window_start = 0
window_end = window_start + window_size

# Select data for the 4-second window
data_window = data[:, window_start:window_end]
times_window = times[window_start:window_end]
df = pd.DataFrame(data.T, columns=channel_names)

# Display the first few rows of the DataFrame
print(df.head())

# Plot the first 5 channels
plt.figure(figsize=(8, 5), dpi=130)
colors = ['b', 'g', 'r', 'c', 'm']  # Different colors for each channel
for i in range(5):
    plt.subplot(5, 1, i + 1)
    
    # Plot the first half of the original signal (up to 600 samples)
    plt.plot(times[:400], data[i, :400], color=colors[i % len(colors)], alpha=0.3, label=f"Full Signal (First Half)")
    
    plt.title(f"Channel: {channel_names[i]}", fontsize=10, fontweight='bold')
    
    if i == 4:
        plt.xlabel("Time (s)", fontsize=9, fontweight='bold')
    if i == 2:  # Add amplitude label only to the middle plot
        plt.ylabel("Amplitude", fontsize=9, fontweight='bold')
    plt.legend(loc="upper left", fontsize=7, frameon=False)

plt.tight_layout()
plt.show()

```
